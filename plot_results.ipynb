```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder = '/home/isacar/workspace/HowToTrainYourMAMLPytorch/'
rows = []
cols = ['experiment', 'seed', 'test_accuracy']
for trial in os.listdir(folder):
    if 'mini' in trial:
        continue
    if not (trial.startswith('sca') or trial.startswith('maml')):
        continue
    experiment = trial[:-2]
    seed = int(trial[-1])
    try:
        logs_path = os.path.join(folder, trial, 'logs')
        test_path = os.path.join(logs_path, 'test_summary.csv')
        test_df = pd.read_csv(test_path)
        x = test_df.test_accuracy_mean.max()
        rows.append([experiment, seed, x])
    except FileNotFoundError as e:
        pass
df = pd.DataFrame(data=rows, columns=cols)
```

```python
sec_per_min = 60
min_per_hour = 60
rows = []
cols = ['experiment', 'seconds']
for trial in os.listdir(folder):
    if not (trial.startswith('sca') or trial.startswith('maml')):
        continue
    df = pd.read_csv(trial + '/logs/summary_statistics.csv')
    seconds = df.epoch_run_time.sum()
    minutes_tot = seconds / sec_per_min
    hours = np.floor(minutes_tot / min_per_hour)
    minutes = minutes_tot - hours * min_per_hour
    rows.append([trial[:-2], seconds])
df = pd.DataFrame(data=rows, columns=cols)
```

```python
df['hours'] = df.seconds / sec_per_min / min_per_hour
df.groupby('experiment').mean()
```
