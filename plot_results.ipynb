```python
import os

import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder = '/home/isacar/workspace/HowToTrainYourMAMLPytorch/'
for trial in os.listdir(folder):
    if not (trial.startswith('sca') or trial.startswith('maml')):
        continue
    print(trial)
    logs_path = os.path.join(folder, trial, 'logs')
    train_val_path = os.path.join(logs_path, 'summary_statistics.csv')
    train_val_df = pd.read_csv(train_val_path)
    train_val_df.plot('epoch', 'val_accuracy_mean')
    try:
        test_path = os.path.join(logs_path, 'test_summary.csv')
        test_df = pd.read_csv(test_path)
        x = test_df.test_accuracy_mean.max()
        plt.hlines(x, train_val_df.epoch.min(), train_val_df.epoch.max(), label='test_score')
        print('Test score:', x)
    except:
        pass
    plt.legend()
    plt.show()
```
